```python
import pandas as pd
```

```python
pd.options.display.float_format = '{:,.0f}'.format
```

```python
orders = pd.read_csv('Pandas_Public_Orders.csv', parse_dates=['Order_Date'])
orders
```

```python
categories = pd.read_csv('Pandas_Public_Categories.csv')
categories
```

```python
# comment/delete it if it is redundant
orders["Year"] = orders["Order_Date"].dt.year
orders["Month"] = orders["Order_Date"].dt.month
orders
```

```python
orders = orders.merge(categories, how="left", left_on="Category", right_on="Category_Name")
orders
```

```python
orders.sort_values("Category_ID", inplace=True) 
# you can use orders.sort_values(["Category_ID", "Category"], inplace=True) if there will be categories in Orders file that
# are missing in Categories file - they will be sorted by name on the end of the data frame
orders.drop(["Category_Name", "Category_ID"], 1, inplace=True)
orders.reset_index(drop=True, inplace=True)
orders
```

```python
orders.to_html("Pandas_Html.html")
```
